```python
# Import required Qiskit modules
from qiskit import QuantumCircuit  # For creating quantum circuits
from qiskit.quantum_info import Statevector  # For computing quantum states
from qiskit.visualization import plot_bloch_multivector  # For visualizing quantum states
```

```python
# Create a quantum circuit with 1 qubit
qc = QuantumCircuit(1)
inital_state = [0, 1]  # Define the initial state |1>
qc.initialize(inital_state, 0)  # Initialize qubit 0 to |1>
# Apply Pauli X gate (NOT gate) to qubit 0
qc.x(0)
```

```python
# Display the circuit diagram in text format
qc.draw()
```

```python
# Draw the circuit diagram using Matplotlib (graphical format)
qc.draw('mpl')
```

```python
# Get the quantum state resulting from the circuit
state = Statevector.from_instruction(qc)
# Visualize the quantum state on the Bloch sphere
plot_bloch_multivector(state)
```
